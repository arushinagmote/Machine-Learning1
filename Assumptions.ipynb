```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
```

```python
data=pd.read_csv('ad.csv',index_col='Unnamed: 0')
```

```python
data
```

```python
x= data.drop(['Sales'],axis=1)
```

```python
x
```

```python
y= data['Sales']
```

```python
y
```

```python
import statsmodels.api as sm
```

```python
X=sm.add_constant(x)
```

```python
X
```

```python
lin_reg=sm.OLS(y,X).fit()
```

```python
lin_reg.summary()
```
